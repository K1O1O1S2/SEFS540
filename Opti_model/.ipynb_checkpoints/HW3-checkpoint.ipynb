```python
from pulp import *
import pandas as pd 
import numpy as np 
```

```python
# Import files 
adjacency = np.loadtxt('Adjacency_HW3.txt', delimiter = ",")
parcels = np.loadtxt('Parcels_HW3.txt', delimiter = ",")
adjacency = adjacency.astype(int)
```

```python
# Data definition 


B = 1000000
m = 121 #Big Number !! just need to be big enough 

# Parcels 
parcel = parcels[:,0]
n_parcels = len(parcel) 
# Cost Matrix 
cost_matrix = parcels[:,2]

# Area Matrix 
area_matrix = parcels[:,1]

# Perimeter Matrix 
p_matrix = parcels[:,3]

# Parcels i adjacent to j 
adj1 = adjacency[:,0]

# Parcels j adjacent to i 
adj2 = adjacency[:,1]

cb = adjacency[:,2]
```

```python
# Model initialization
model_p= LpProblem("HW3_p", LpMinimize)
#model_a= LpProblem("HW3_a", LpMaximize)
```

```python
# Decision variables 
variable_names = [str(i) for i in range(1,n_parcels+1)]

DV_variables = LpVariable.matrix("X", variable_names, cat = "Binary", lowBound= 0 )
X_allocation = np.array(DV_variables).reshape(100)

# Flow variables
adj_variable_names = [str(adj1[i]) + '_' + str(adj2[j]) for i in range (len(adj1)) for j in range (len(adj2))]
Flow_variables = LpVariable.matrix("Y", adj_variable_names, cat = "Binary", lowBound = 0)
Yij_allocation = np.array(Flow_variables).reshape(260,260)

#Tail variables 

Zij_variable_names = [str(adj1[i]) + '_' + str(adj2[i]) for i in range (len(adj1))]
Zij_variables = LpVariable.matrix("Z", Zij_variable_names, cat = "Integer", lowBound = 0)
Zij_allocation = np.array(Zij_variables).reshape(260)

W_variable_names = [str(i) for i in range(1,n_parcels+1)]
W_variables = LpVariable.matrix("W", W_variable_names, cat = "Integer", lowBound= 0 )
W_allocation = np.array(W_variables).reshape(100)

# Indicator variable 
U_variable_names = [str(adj1[i]) + '_' + str(adj2[j]) for i in range (len(adj1)) for j in range(len(adj2))]
Flow_variables = LpVariable.matrix("U", U_variable_names, cat = "Binary", lowBound = 0)
Uij_allocation = np.array(Flow_variables).reshape(260,260)

Budget = LpVariable("Budget", cat="Continuous", lowBound=0)
AREA = LpVariable("AREA", cat="Continuous", lowBound=0)
```

```python
# Constraints 

## Constraint 1: Budget constraint 
### SUM pi*Xi <= B 

## Constraint 2: 4 adjacent cells 
### SUM Yij (i e Ij)<= [|hdhdhd]*Xj for all j 

## Constraint 3: Max 1 connection from i
#### SUM Yij (j e Ii) <= Xi 

## Constraint 4: Number of arcs must be less or equal to the number of nodes 
#### SUM Yij (i,j) <= SUM Xi - 1 (i)

## Constraint 5: tail function 
### Zij >= Wi + 1 - m(1-Yij) for all i, j adjacent, i != j 

## Constraint 6: cycle breaking constraint 
### Wj >= SUM (ieIj) Zij for all j 

## Constraint 7: ensure that zij is minimal 
### Zij <= m*Yij 

## Constraint 8:ensure that zij is minimal
### Zij <= Wi+1 

#core parcel :  parcel 10, 32.031, 0.00
len(adj2)
```

```python
## Constraint 1: Budget constraint 
### SUM pi*Xi <= B 

#model_a += lpSum(X_allocation*cost_matrix) <= B, "Budget constraint" 
model_p += lpSum(X_allocation*area_matrix) >= 290.62, "Area constraint" 
```

```python
## Constraint 2: 4 adjacent cells 
### SUM Yij (i e Ij)<= 4*Xj for all j  !!! change to cardinality of Aj number of parcels adjacent to j 
count = [] 
c = 0 
for j in range(n_parcels):
    for k in range(len(adj2)): 
        if adj2[k] == parcel[j]:
            count.append(k) 
            c += 1 
            #print(count)
            #print(adj2[k])
            #print(parcel[j])
    if count != []:
        #print(lpSum(Yij_allocation[count[i]][count[i]] for i in range(len(count)) ) <= c*X_allocation[j])
        #model_a += lpSum(Yij_allocation[count[i]][count[i]] for i in range(len(count))) <= c*X_allocation[j], "Max connection to j" + str(j)
        model_p += lpSum(Yij_allocation[count[i]][count[i]] for i in range(len(count))) <= c*X_allocation[j], "Max connection to j" + str(j)

    count = []
    c = 0 
```

```python
Uij_allocation
```

```python
## Constraint 3: Max 1 connection from i
#### SUM Yij (j e Ii) <= Xi 

count = [] 
for i in range(n_parcels):
    for k in range(len(adj1)): 
        if adj1[k] == parcel[i]:
            count.append(k) 
            #print(count)
            #print(adj2[k])
            #print(parcel[j])
    if count != []:
        #print(lpSum(Yij_allocation[count[i]][count[i]] for i in range(len(count)) ) <= X_allocation[i])
        #model_a += lpSum(Yij_allocation[count[j]][count[j]] for j in range(len(count))) <= X_allocation[i], "Max connection from i" + str(i)
        model_p += lpSum(Yij_allocation[count[j]][count[j]] for j in range(len(count))) <= X_allocation[i], "Max connection from i" + str(i)

    count = []
```

```python
## Constraint 4: Number of arcs must be less or equal to the number of nodes 
#model_a += lpSum(Yij_allocation[j][j] for j in range(len(Yij_allocation))) == lpSum(X_allocation[i] for i in range(0,100))-1, "Number of arcs"
model_p += lpSum(Yij_allocation[j][j] for j in range(len(Yij_allocation))) == lpSum(X_allocation[i] for i in range(0,100))-1, "Number of arcs" 
```

```python
## Constraint 5: Number of arcs must be less or equal to the number of nodes 
### Zij >= Wi + 1 - m(1-Yij) for all i, j adjacent, i != j 

for i in range(n_parcels):
    for k in range(len(adj1)):
        if adj1[k] == i :
            count.append(k)
    if count != []:
        for j in range(len(count)):
            Zij_allocation[count[j]] >= W_allocation[i-1] + 1 - m*(1-Yij_allocation[count[j]][count[j]])
            #model_a += Zij_allocation[count[j]] >= W_allocation[i-1] + 1 - m*(1-Yij_allocation[count[j]][count[j]]), "Tail function" + str(i) + '_'+ str(j)
            model_p += Zij_allocation[count[j]] >= W_allocation[i-1] + 1 - m*(1-Yij_allocation[count[j]][count[j]]), "Tail function" + str(i) + '_'+ str(j)

    count = []
```

```python
## Constraint 6: Cycle breaking constraint 
### Wj >= SUM (ieIj) Zij for all j 

count = []
for i in range(n_parcels):
    for j in range(len(adj2)):
        if adj2[j] == i: 
            count.append(j)
    if count != []:
        W_allocation[i-1] == lpSum(Zij_allocation[count[k]] for k in range(len(count)) )
        #model_a += W_allocation[i-1] == lpSum(Zij_allocation[count[k]] for k in range(len(count)) ), "Loop_break function" + str(i)
        model_p += W_allocation[i-1] == lpSum(Zij_allocation[count[k]] for k in range(len(count)) ), "Loop_break function" + str(i)

    count = []
```

```python
## Consraint 7: Ensure that zij is minimal 
### Zij <= m*Yij 
count = []
ab = 0 
for i in range(n_parcels): 
    for j in range(len(adj2)): 
        if adj2[j] == i: 
            count.append(j)
    if count != []: 
        for k in range(len(count)): 
            Zij_allocation[count[k]] <= m*Yij_allocation[count[k]][count[k]]
            ab = ab+1
            #model_a += Zij_allocation[count[k]] <= m*Yij_allocation[count[k]][count[k]], "Ensure_min_Zij" +  str(ab)
            
    count = []
```

```python
## Constraint 8: Ensure that zij is minimal 
### Zij <= Wi+1 
count = []
ab = 0 
a = 0 
for i in range(n_parcels): 
    for j in range(len(adj1)): 
        if adj1[j] == i: 
            count.append(j)
    if count != []:
        a = a + 1
        for k in range(len(count)): 
            Zij_allocation[count[k]] <= W_allocation[i-1]+1
            ab = ab+1
            #model_a += Zij_allocation[count[k]] <= W_allocation[i-1]+1, "Ensure_min_Zij_2" + str(ab)
            
    count = []
```

```python
#core parcel : parcel 10, 32.031, 0.00
#model_a += X_allocation[22] == 1, "Core parcel"
model_p += X_allocation[22] == 1, "Core parcel"
```

```python
# Ui constraints 

#### 2*Zm <= Sum Xm,t + Sum Ym,t 
for i in range (n_parcels):
    for k in range(len(adj1)):
        if adj1[k] == parcel[i]:
            a = adj2[k]
            parcel_ind = np.where(parcel ==a)
            model_p += (X_allocation[i] + X_allocation[parcel_ind]
                        >= 2*Uij_allocation[k][k], "Uij_con1_" + str(i)+"_"+str(a))      
#### Sum Ym,t + Sum Xm,t - Zm <= 1 
for i in range (n_parcels):
    for k in range(len(adj1)):
        if adj1[k] == parcel[i]:
            a = adj2[k]
            parcel_ind = np.where(parcel ==a)
            model_p += (X_allocation[i] + X_allocation[parcel_ind]
                        - Uij_allocation[k][k] <=1, "Uij_con2_" + str(i)+"_"+str(a)) 
```

```python
# budget 
model_p += lpSum(X_allocation*cost_matrix)==Budget, "BUDGET"
model_p += lpSum(X_allocation*area_matrix)==AREA, "Area"
```

```python
# Objective function 

#obj_func = lpSum(X_allocation*area_matrix)
#model_a += obj_func
obj_func_p = lpSum(X_allocation*p_matrix) -lpSum(Uij_allocation[k][k]*cb[k] for k in range(len(Uij_allocation)))
model_p += obj_func_p
#print(model)

```
